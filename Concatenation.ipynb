```python
data_list=[354972062885738, 354972062885746,354972062885753,354972062886561,354972062886587,354972062886595,354972062886611]
```

```python
df_list=[]
for each in data_list:
    df="Data_"+str(each)+".csv"
    df_list.append(df)
df_list  
```

```python
import pandas as pd
dataframes_list=[]
for each in df_list:
    dataframes_list.append(pd.read_csv(each))
```

```python
import numpy as np
lst=[]
for each in dataframes_list:
    print(each.shape)
    lst.append(each.shape)
np.sum(pair[0] for pair in lst)
```

```python
Full_data = pd.concat(list(dataframes_list))
Full_data.to_csv("Full_data.csv")
```
