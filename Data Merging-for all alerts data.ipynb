```python
import time
start_time = time.time()
import pandas as pd
import numpy as np
from datetime import datetime
# from datetime import datetime
data1= pd.read_csv(r"C:\Users\INFO-DSK-04\Desktop\Asset_Data\SQL Server\Data_354972062885738.csv")
data2= pd.read_csv(r"C:\Users\INFO-DSK-04\Desktop\Asset_Data\SQL Server\Data_354972062885738_with_alerts.csv")
data2.drop("TrackDateTime",axis=1,inplace=True)
data2.rename(columns={"TrackDateTime_new":"TrackDateTime"},inplace=True)
```

```python
#Function for changing the syntax of datetime to a paticular format
new_format = "%Y:%m:%d %H:%M:%S"
def change_datetime_syntax(date_text):
    try:
        d=datetime.strptime( date_text , '%d-%m-%Y %H:%M:%S.%f') 
        date_text=d.strftime(new_format)
        return date_text
    except:  
        try:
            d=datetime.strptime(date_text , '%d-%m-%Y %H:%M:%S') 
            date_text=d.strftime(new_format)
            return date_text
        except:
            try:
                d=datetime.strptime(date_text , '%Y:%m:%d %H:%M:%S.%f') 
                date_text=d.strftime(new_format)
                return date_text
            except:
                try:
                    d=datetime.strptime(date_text , '%Y:%m:%d %H:%M:%S') 
                    date_text=d.strftime(new_format)
                    return date_text
                except:
                    try:
                        d=datetime.strptime(date_text , '%Y-%m-%d %H:%M:%S.%f') 
                        date_text=d.strftime(new_format)
                        return date_text
                    except:
                        try:
                            d=datetime.strptime(date_text , '%Y-%m-%d %H:%M:%S') 
                            date_text=d.strftime(new_format)
                            return date_text
                        except:
                            print(date_text,"New format found, include this in function")
```

```python
data1['TrackDateTime_new']=[change_datetime_syntax(k) for k in data1.TrackDateTime]
data2['TrackDateTime_new']=[change_datetime_syntax(k) for k in data2.TrackDateTime]
```

```python
Full_data = pd.concat([data1,data2], ignore_index=True, sort=False).sort_values(['TrackDateTime_new'])
# Full_data['TrackDateTime'] = datetime.fromtimestamp(Full_data['TrackDateTime'])
Full_data["AlertTypeName"]=Full_data.AlertType
Full_data=Full_data.replace({'AlertTypeName': {32: "Rash Turning", 14: "Harsh Breaking",15: "Harsh Acceleration"}})
Full_data.reset_index(drop=True,inplace=True)
Full_data=Full_data[['AssetNo', 'DeviceNo', 'Latitude', 'Longitude', 'Location','TrackDateTime_new','Altitude', 'Speed','AlertType','AlertTypeName', 'Direction']]

# Removing duplicate rows with or without alerts.
Full_data_new=Full_data.groupby(by='TrackDateTime_new').agg({'AssetNo':'first',
                                           'DeviceNo' :'max',
                                           'Latitude'  :'max',
                                           'Longitude'  :'max',
                                           'Location' : 'first',
                                           'Speed': 'max',
                                           'AlertType': 'max',
                                            'AlertTypeName':'first',
                                           'Altitude': 'mean'}).reset_index()
Full_data_new.AlertType.replace(0, np.nan, inplace=True)
# Full_data_new=Full_data_new.sort_values([ 'TrackDateTime'])
Full_data_new.to_csv(r"C:\Users\INFO-DSK-04\Desktop\Asset_Data\Data_354972062885738.csv",index=False)
print("--- %s seconds ---" % (time.time() - start_time))
```
